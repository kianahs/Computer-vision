```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def horizontally_skew_image(image, skew_angle):
    height, width = image.shape[:2]

    # Calculate the tangent of the skew angle
    tan_theta = np.tan(np.deg2rad(skew_angle))

    # Define the transformation matrix
    M = np.float32([[1, tan_theta, 0],
                    [0, 1, 0]])

    # Apply the transformation to the image
    skewed_image = cv2.warpAffine(image, M, (width, height))

    return skewed_image

# Load the input image
input_image = cv2.imread('test.jpg')

# Set the skew angle (positive values skew to the right, negative values skew to the left)
skew_angle = 10

# Skew the image
skewed_image = horizontally_skew_image(input_image, skew_angle)

# Display the original and skewed images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
axes[1].imshow(cv2.cvtColor(skewed_image, cv2.COLOR_BGR2RGB))
axes[1].set_title('Skewed Image')
plt.tight_layout()
plt.show()
```
